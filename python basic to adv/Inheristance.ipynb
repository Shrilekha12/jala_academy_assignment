```python
A, B and C are classes 
A is a super class. B is a sub class of A. C is a sub class of B.  
Create three methods in each class, 2 methods are specific to each class and third 
method (override method) should be in all three Classes A, B and C 
Create a class with main method. Create an object for each class A, B and C in main 
method and call every method of each class using its own object/instance. 
Call an overridden method with super class reference to B and C classâ€™s objects 
Runtime Polymorphism with Data Members/Instance variables, Repeat the above 
process only for data members
```

```python
# Superclass A
class A:
    def __init__(self):
        self.data = "Data from class A"
        
    def method1_A(self):
        print("Method 1 specific to class A")
        
    def method2_A(self):
        print("Method 2 specific to class A")
        
    def override_method(self):
        print("Override method in class A")
        
    def show_data(self):
        print("Class A data:", self.data)

# Subclass B inherits from A
class B(A):
    def __init__(self):
        super().__init__()
        self.data = "Data from class B"
        
    def method1_B(self):
        print("Method 1 specific to class B")
        
    def method2_B(self):
        print("Method 2 specific to class B")
        
    def override_method(self):
        print("Override method in class B")
        
    def show_data(self):
        print("Class B data:", self.data)

# Subclass C inherits from B
class C(B):
    def __init__(self):
        super().__init__()
        self.data = "Data from class C"
        
    def method1_C(self):
        print("Method 1 specific to class C")
        
    def method2_C(self):
        print("Method 2 specific to class C")
        
    def override_method(self):
        print("Override method in class C")
        
    def show_data(self):
        print("Class C data:", self.data)

# Main class to run the tests
class MainClass:
    @staticmethod
    def main():
        print("=== Creating instances ===")
        a = A()
        b = B()
        c = C()

        print("\n=== Calling methods of class A ===")
        a.method1_A()
        a.method2_A()
        a.override_method()

        print("\n=== Calling methods of class B ===")
        b.method1_B()
        b.method2_B()
        b.override_method()

        print("\n=== Calling methods of class C ===")
        c.method1_C()
        c.method2_C()
        c.override_method()

        # Runtime Polymorphism: superclass reference to subclass objects
        print("\n=== Runtime Polymorphism: calling override_method using superclass reference ===")

        # Using class A reference to B and C objects
        a_ref = b
        print("Calling override_method on A reference to B object:")
        a_ref.override_method()  # Should call B's override_method

        a_ref = c
        print("Calling override_method on A reference to C object:")
        a_ref.override_method()  # Should call C's override_method

        # Demonstrate polymorphism with data members
        print("\n=== Polymorphism with data members ===")

        print("Using class A instance:")
        a.show_data()

        print("Using class B instance:")
        b.show_data()

        print("Using class C instance:")
        c.show_data()

        # Using superclass reference to subclass objects for data members
        a_ref = b
        print("Data accessed via A reference to B object:")
        a_ref.show_data()

        a_ref = c
        print("Data accessed via A reference to C object:")
        a_ref.show_data()

if __name__ == "__main__":
    MainClass.main()
```
